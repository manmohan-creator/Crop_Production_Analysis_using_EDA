```python
plt.figure(figsize = (10,5))
plt.bar(bottom10.keys(), bottom10.values, color = 'blue', edgecolor = 'pink')
plt.title("Bottom 10 district counts", fontdict = {"size":12})
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.bar(top10.keys(), top10.values, color = 'green', edgecolor = 'pink')
plt.title("Top 10 district counts", fontdict = {"size":12})
plt.xticks(rotation = 90)
plt.show()
```
